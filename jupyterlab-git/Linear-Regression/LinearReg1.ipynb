```python
import numpy as np 
import matplotlib.pyplot as plt 

plt.title('Correlation Between Distance and Velocity of Galaxies', fontsize=15, color = "darkblue")
  
def estimate_coef(x, y): 
    # observation quantity 
    n = np.size(x) 
  
    # mean of x & y 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # cross-deviation and deviation about x 
    SS_xy = np.sum(y*x - n*m_y*m_x) 
    SS_xx = np.sum(x*x - n*m_x*m_x) 
  
    # regression coefficients 
    b1 = SS_xy / SS_xx 
    b0 = m_y - b1*m_x 
  
    return(b0, b1) 
  
def plot_regression_line(x, y, z): 
    # points on graph  
    plt.scatter(x, y, color = "brown", 
               marker = "o", s = 30) 
  
    # predicted response 
    #vector 
    y_pred = z[0] + z[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "darkblue") 
  
    # putting labels 
    plt.xlabel('Distance From Earth') 
    plt.ylabel('Velocity of Galaxy') 
  
    # function to show plot 
    plt.show() 
  
def main(): 
    # values  
    x = np.array([458.7383824,
272.2113088,
642.9251471,
168.6132353,
279.4472206,
150.8315294,
619.7135294,
369.8401765,
241.8624706,
417.0415588,
215.6313529,
180.4677794,
140.0675735,
348.0456176,
546.3754412,
176.7952941,
636.4394118,
263.1055588,
651.1142647,
326.2874559,
210.5015735,
384.0281471,
243.5962059,
404.5873676,
193.0980441,
293.5955735,
420.8379265,
446.43,
642.6330882,
503.5592647,
558.9569118,
481.4064706,
364.62075,
267.4977353,
602.2213235,
397.2633971,
569.6867647,
322.0952206,
513.8788235,
200.5232647,
463.4245588,
581.4904412,
492.9061765,
424.1061176,
588.9944118,
335.76075,
310.4376176,
561.1023529,
478.9570588,
530.3439706]) 
    y = np.array([34881.20217,
19366.82027,
42678.62057,
12562.16838,
19843.73709,
11047.05563,
37114.51493,
25050.46321,
17428.83958,
36939.98907,
18202.33401,
12211.46845,
12002.93996,
24798.98191,
36550.39658,
13373.20881,
37394.82465,
18979.88478,
37677.66644,
22757.68686,
15676.97806,
26336.41625,
15507.47855,
26546.14597,
14054.97586,
20947.30274,
27191.4632,
30688.11153,
42084.99968,
30883.47287,
36791.00519,
33519.03604,
25787.83397,
18918.16386,
39470.78391,
27180.76013,
36031.52966,
23123.55992,
33905.47515,
14973.08765,
35900.87161,
38209.54771,
30486.61163,
29652.71061,
38960.4771,
28771.48444,
22217.70702,
33007.6342,
29860.8199,
32973.15268]) 
  
    # estimated coefficients 
    b = estimate_coef(x, y) 
    print("Estimated coefficients:\nb_0 = \ nb_1 = {}".format(b[0], b[1])) 
  
    # plotting slope line 
    plot_regression_line(x, y, b) 
   
  #output 
if __name__ == "__main__": 
    main()
```
